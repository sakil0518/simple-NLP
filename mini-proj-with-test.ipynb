```python
# importing all neccessary packages

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# user input for wanted dataset
print("Enter file name of dataset:")
dataset = input(str())
df=pd.read_csv(dataset)

# true class labels
y = df[df.columns[-1]]

# x is instances without true class
df = df.drop(df.columns[-1], axis=1)
x = np.array(df)

# change categorical values to numerical
le = preprocessing.LabelEncoder()
for i in range(len(x)-2):
    x[:,i] = le.fit_transform(x[:, i])
y = le.fit_transform(y)
```

```python
# create decision tree classifier using entropy as criterion for splitting
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
# split data to training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6)
```

```python
# fit training data to our decisiontree classifier and show
clf.fit(x_train,y_train)
plot_tree(clf)
```

```python
# use classifier to predict class for prediction set
y_pred = clf.predict(x_test)

# how well is our decision classifier performing
print(classification_report(y_test, y_pred))
```
